```python
import json
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## H2

### Plot 1

```python
filename = "E_H2_shots=1024_N_0=10_N_exp=10000_T_060225_171018.json"
with open(filename, 'r') as f:
    data_10 = json.load(f)
# print(data_10)

sns.kdeplot(data_10['uniform'], linestyle='--', label='H2 Uniform')
sns.kdeplot(data_10['vmsa'], linestyle='--', label='H2 VPSR')
sns.kdeplot(data_10['vpsr'], linestyle='--', label='H2 VMSA')
plt.axvline(data_10['statevector'], linestyle='dotted', label='Statevector (Exact) Simulation', color='black')
plt.legend()
```

```python
filename = "E_H2_shots=1024_N_0=50_N_exp=10000_T_060225_185018.json"
with open(filename, 'r') as f:
    data_50 = json.load(f)
# print(data_50)

sns.kdeplot(data_50['uniform'], linestyle='--', label='H2 Uniform')
sns.kdeplot(data_50['vmsa'], linestyle='--', label='H2 VPSR')
sns.kdeplot(data_50['vpsr'], linestyle='--', label='H2 VMSA')
plt.axvline(data_50['statevector'], linestyle='dotted', label='Statevector (Exact) Simulation', color='black')
plt.legend()
```

```python
filename = "E_H2_shots=1024_N_0=100_N_exp=10000_T_060225_204155.json"
with open(filename, 'r') as f:
    data_100 = json.load(f)
# print(data_100)

sns.kdeplot(data_100['uniform'], linestyle='--', label='H2 Uniform')
sns.kdeplot(data_100['vmsa'], linestyle='--', label='H2 VPSR')
sns.kdeplot(data_100['vpsr'], linestyle='--', label='H2 VMSA')
plt.axvline(data_100['statevector'], linestyle='dotted', label='Statevector (Exact) Simulation', color='black')
plt.legend()
```

### Plot 2

##### H2 VPSR

```python
x_lim = 0.025

data_10_normalized = np.array(data_10['vpsr']) - data_100['statevector']
data_50_normalized = np.array(data_50['vpsr']) - data_100['statevector']
data_100_normalized = np.array(data_100['vpsr']) - data_100['statevector']

sns.kdeplot(data_10_normalized, linestyle='--', label='H2 VPSR N0 = 10')
sns.kdeplot(data_50_normalized, linestyle='--', label='H2 VPSR N0 = 50')
sns.kdeplot(data_100_normalized, linestyle='--', label='H2 VPSR N0 = 100')

plt.axvline(0, linestyle='dotted', label='Statevector (Exact)', color='black')
plt.xlim(-x_lim, x_lim)
plt.legend(loc='upper left', frameon=False)

plt.xlabel("Estimated Energy", fontsize=13)
plt.ylabel("Density", fontsize=13)

plt.savefig("H2_Bench_VPSR.pdf", format="pdf", dpi=300, bbox_inches="tight")
```
